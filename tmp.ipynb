```python
import torch
import numpy
import pandas as pd
import sklearn
print(torch.__version__)
print(torch.cuda.is_available())
print(f"Using Scikit-Learn version: {sklearn.__version__} (materials in this guide require this version or newer).")
```

```python
data=[[1,2],[3,4]]
x_data=torch.tensor(data)
```

```python
print(x_data.device)
shape=(2,3,)
rt=torch.rand(shape)
print(rt)
print(rt.dtype)
```

```python
t1=torch.cat([rt,rt,rt],dim=1)#default dim=0
print(t1)
```

```python
#matrix multiplication
y1=rt@rt.T
y2=rt.matmul(rt.T)
y3 = torch.rand(2,2)
torch.matmul(rt,rt.T,out=y3)
print(y1,y2,y3)
#element-wise
z1=rt*rt
z2=rt.mul(rt)
z3 = torch.rand_like(z1)
torch.mul(rt,rt,out=z3)
print(z1,z2,z3)
```

```python
#Single-element tensors
sum=y1.sum()
si=sum.item()
print(sum, type(sum),si,type(si))
```

```python
#In-place operations _
print(y1.add_(5))
```

```python
# numpy change with tensor
n1=y1.numpy()
print(n1)
y1.add_(-5)
print(n1)
```

```python
# tensor change with numpy
np=numpy.ones((3,4))
t2=torch.from_numpy(np)
numpy.add(np,1,out=np)
print(np)
print(t2)
```
